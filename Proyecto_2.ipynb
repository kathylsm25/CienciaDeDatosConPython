```python
# Creamos nuestro Dataframe
df_autos = pd.read_csv('autos.csv', sep=';') # sep=';'
df_autos.head()
```

```python
'''
Creamos una función para recodificar el País de Origen
  Cuando es 'Europe' cambia a 'Panama'
            'USA' cambia a Australia
            'Japan' cambia a Italia
'''

def recodifica_pais_origen  (pais_origen):
  '''
  Esta función recodifica el país de origen de nuestro DataFrame. Actualmente tenemos 3 País de Origen en nuestra data (Europe, Japan, USA). 
  La función reemplaza estos países por Panamá, Italia y Australia
  '''
  if pais_origen == "Europe":
      return "Panama"
  elif pais_origen == "Japan":
     return "Italia"
  elif pais_origen == "USA":
     return "Australia"
  else: return "País Desconocido" 
```

# Ejercicio 3
Crear una función que permita generar 3 variables cualitativas
ordinales a partir de las variables numéricas con los siguientes criterios:
★ Si es mayor a la “media + 1*desviación estándar” → Alto
★ Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
★ Si es menor a la “media - 1*desviación estándar” → Bajo

Nota: Mostrar con las nuevas variables generadas.
Elegir con libertad las variables a trabajar.
La función debe contar con su respectiva documentación.

```python
# Obtenemos la sintáxis de como crear una función
# def nombre(lista de parámetros):         # parámetros de la función
#   """Documentación"""
#   sentencias
#   return [expresion]
# La Documentación de la función será de la siguiente manera:

#   Si x > (media - (1 * desviacion_std)) = 'Alto'
#   Si x > (media - (1 * desviacion_std)) and x < (media + (1 * desviacion_std)) = 'Medio'
#   Si x < (media - (1 * desviacion_std)) = 'Bajo'
  
# '''

def func_generar_variables_cualitativas_ordinales(var_num):
    ''' Esta Función recibe el valor de una variable cuantitativa de nuestro DataFrame de Autos y genera una columna adicional con los siguientes variables cualitativas

      Si var_num > (media + (1 * desviacion_std)) = 'Alto'
      Si var_num > (media - (1 * desviacion_std)) and x < (media + (1 * desviacion_std)) = 'Medio'
      Si var_num < (media - (1 * desviacion_std)) = 'Bajo'

      Para la condicional utilizamos la primera y la última y todo lo que no entre dentro de estas se le asignará Medio

      Argumentos:
      Variable_Cuantitativa: Introduzca la variable numérica con la cual desea que se evaluen para clasificarlo como Alto, Medio o Bajo

      Retorna:
      Devuelte las variables Alto, Medio, Bajo dependiendo del valor que tenga el Desplazamiento

    '''

    media = np.mean(var_num)
    desviacion_std = np.std(var_num)
    variables_cualitativas = []
    for x in var_num:
        if x > (media + (1 * desviacion_std)):
            variables_cualitativas.append('Alto')
#        elif x < (media - (1 * desviacion_std)):
#            variables_cualitativas.append('Bajo')
        elif x > (media - (1 * desviacion_std)) and x < (media + (1 * desviacion_std)):
            variables_cualitativas.append('Medio')
        else:
            variables_cualitativas.append('Bajo')
    return variables_cualitativas
```

```python
# Obtenemos la media y desviación estándar de la variable Desplazamiento para comprobar nuestros resultados
media = np.mean(df_autos['Desplazamiento'])
desviacion_std = np.std(df_autos['Desplazamiento'])
print ('La media es', media , ' y la desviación estándar es',desviacion_std)
```

```python
# Probamos nuestra función con la variable cuantitativa de Desplazamiento

df_autos = df_autos.assign(variable_6 = func_generar_variables_cualitativas_ordinales(df_autos['Desplazamiento']))



df_variable_6=df_autos.loc[:,['Desplazamiento', 'variable_6']]
df_variable_6

```

```python
# Probamos nuestra función con otra variable cuantitativa Caballo_De_Fuerza
df_autos = df_autos.assign(variable_7 = func_generar_variables_cualitativas_ordinales(df_autos['Caballo_De_Fuerza']))



df_variable_7=df_autos.loc[:,['Caballo_De_Fuerza', 'variable_7']]
df_variable_7
```

```python
# Importamos librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
import missingno as msno
```

```python
# Creamos nuestra función y la documentamos

def func_graficas(df, var_cuali, var_num):
    """
    Esta función muestra tres gráficos simultáneamente para una variable numérica y una variable cualitativa:
    cajas y bigotes, barras de la media y histograma con densidad (curva).
    
    Parámetros:
    - df: DataFrame que contiene los datos.
    - var_num: Nombre de la variable numérica.
    - var_cuali: Nombre de la variable cualitativa.

    Retorna: 
    Grafico de Caja y Bigotes
    Grafico de Barras
    Grafico de Histogramas (Curva)
    """
    
    # Gráfico de cajas y bigotes
    sns.boxplot(x=var_cuali, y=var_num, data=df, palette = 'Greens')
    plt.title('Cajas y bigotes')
    plt.show()
    
    # Gráfico de barras de la media
    sns.barplot(x=var_cuali,y=var_num,  data=df, estimator = np.mean, palette='Purples')
    plt.title('Barras de la media')
    plt.show()
    
    # Gráfico de histograma con densidad
    sns.histplot(x=var_num, hue=var_cuali, data=df, kde=True, palette = 'colorblind')
    plt.title('Histograma con densidad')
    plt.show()
```

Crear una función que devuelva un mapa de calor con la correlación de todas las variables numéricas (incluye las generadas), para cada país. 
● La función debe contar con su respectiva documentación.


```python
# Revisamos las variables numéricas de nuestro dataframe
df_autos.info()
```

```python
# Seleccionamos las variables numéricas
var_numericas = df_autos.select_dtypes(include = ['float64','int64','number'])
var_numericas
```

```python
# Calculamos la correlación de estas variables númericas con la variable cualitativa de Pais
correlacion_pais = var_numericas.groupby(df_autos['Pais_Origen']).corr()
correlacion_pais
```

```python
# Generar un mapa de calor con las correlaciones

plt.figure(figsize=(10, 8))
sns.heatmap(correlacion_pais, cmap='coolwarm', annot=True, fmt='.2f', linewidths=1.5)
plt.title('Correlación de variables numéricas por país')
plt.show()
```

```python
# Creamos nuestra función
def func_genera_mapa_calor (data, campo_pais, pais):
  """
    Esta función toma un conjunto de datos (DataFrame), el nombre de la columna que contiene el nombre del país
    y el nombre del país como entrada y devuelve un mapa de calor que muestra la correlación de todas
    las variables numéricas para ese país en particular.
    
    Parámetros:
    data (DataFrame): un dataframe con todas las variables
    columna_pais (str): el nombre de la columna o campo que contiene el nombre del país
    pais (str): el nombre del país para el cual se desea generar el mapa de calor
    
    Retorna:
    Una visualización de un mapa de calor de correlaciones de todas las variables numéricas para el país indicado.
    
  """
  # Verificamos si el país existe en el DataFrame
  if pais not in data[campo_pais].unique():
        print(f"Error: el país '{pais}' no se encuentra en el DataFrame.")
        return

  # Filtramos los datos para obtener el país al cual le queremos generar la correlación
  data_pais = data[data[campo_pais] == pais]

  # Seleccionamos las variables numericas
  variables_numericas = data_pais.select_dtypes(include = ['float64','int64','number'])

  # Calculamos la matriz de correlación del pais con las variables númericas.
  correlacion_pais = variables_numericas.groupby(data[campo_pais]).corr()

  # Generamos el mapa de valor con seaborn. Aplicamos una máscara para el triángulo superior del grafico
  mask = np.triu(np.ones_like(correlacion_pais, dtype=bool ))
  plt.figure(figsize=(10, 8)) # colocamos la base
  sns.heatmap(correlacion_pais, cmap='Purples', annot=True, fmt='.2f', linewidths=1.5, mask = mask,  vmax=1, center = 0, 
            square = True)

  # Añadimos Titulo y las bases de eje x y ya con los nombres de las variables
  plt.title(f"Mapa de calor de correlaciones para {pais}")
  plt.xlabel("Variables")
  plt.ylabel("Variables")

  plt.show()


```

```python
# Ejecutamos nuestra función para probarla
func_genera_mapa_calor(df_autos,'Pais_Origen', 'Australia')
```
